```python
import pandas as pd
```

```python
our_table = pd.read_csv('dataset_with_tones.csv')
our_table.head()
```

```python
reference_table = pd.read_csv('reference.csv')[['Иероглиф', 'Цифровое обозначение тонов']]
reference_table.head()
```

```python
def merge_cols(a, b):
    index_a = [entry[1]['0'] for entry in a.iterrows()]
    index_b = [entry[1]['Иероглиф'] for entry in b.iterrows()]
    a = [entry[1]['5'] for entry in a.iterrows()]

    b = [entry[1]['Цифровое обозначение тонов'] for entry in b.iterrows()]

    present_in_b = []
    for entry in index_b:
        if entry in index_a:
            present_in_b.append(a[index_a.index(entry)])
        else:
            # Логи о пропущенных нами иероглифах
            # print(f'Entry {entry} is not found!')
            present_in_b.append(pd.NaT)
        
    return pd.DataFrame(
        {'Иероглиф': index_b, 'Лингвистика': b, 'Наше решение': present_in_b}
    )
```

```python
result_dataset = merge_cols(our_table, reference_table)
result_dataset
```
