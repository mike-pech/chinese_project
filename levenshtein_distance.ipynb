```python
import pandas as pd
import Levenshtein

data = pd.read_csv('C:\\Users\\xiaom\\Downloads\\quality_assurance.csv')

data_filtered = data.dropna(subset=['Наше решение'])

levenshtein_distances = data_filtered.apply(lambda row: Levenshtein.distance(str(row['Лингвистика']), str(row['Наше решение'])), axis=1)

data['Levenshtein_distance'] = levenshtein_distances

average_levenshtein_distance = data['Levenshtein_distance'].mean()

print("Среднее арифметическое расстояния Левенштейна:", average_levenshtein_distance)

data.to_csv('dataset_levenshtein.csv', index=False)
```
